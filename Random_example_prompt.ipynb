```python
from transformers import (AutoModelForCausalLM,
                          BitsAndBytesConfig,
                          AutoTokenizer,
                          TrainingArguments
                          )
import torch
```

```python
model_name = "NousResearch/Llama-2-7b-hf"
quant_config = BitsAndBytesConfig(use_4bit=True,
                                  bnb_4bit_compute_dtype=torch.bfloat16,
                                  bnb_4bit_quant_type='nf4')


```

```python
PROMPT_TEMPLATE = """{source}: {source_sentence} {dest}: """


```

```python
!pip install datasets

from datasets import Dataset
def build_parallel_ds(source_str, ref_str, PROMPT_TEMPLATE=PROMPT_TEMPLATE):
  ref_list = ref_str.split("\n")
  src_list = source_str.split("\n")
  assert len(ref_list) == len(src_list)
  ds = {"src":[],"ref":[],"prompt":[]}
  for i in range(len(ref_list)):
    ds["src"].append(src_list[i])
    ds["ref"].append(ref_list[i])
    prompt = PROMPT_TEMPLATE.format(
        source="English",
        dest="German",
        source_sentence=src_list[i])
    ds["prompt"].append(prompt)

  return Dataset.from_dict(ds)
```

```python
!pip install unbabel-comet
from comet import download_model, load_from_checkpoint

comet_path = download_model("Unbabel/wmt20-comet-da")
comet_model = load_from_checkpoint(comet_path)
```

```python
import random
```

```python
model_ids = ["haoranxu/ALMA-7B", "mistralai/Mistral-7B-v0.3", "deepseek-ai/deepseek-llm-7b-base", "NousResearch/Llama-2-7b-hf"]
```

```python
source_str = open(src_path).read()
ref_str = open(ref_path).read()
```

```python
ds = build_parallel_ds(source_str, ref_str)
print(ds)
```

```python

def eval_comet_monolingual(parallel_ds, model, tokenizer, comet_model, mode="none"):
    scores = []
    k = random.randint(0, len(parallel_ds) - 2)
    random_src = ""
    random_ref = ""
    if mode == "both":
        random_src = parallel_ds["src"][k] + " "
        random_ref = parallel_ds["ref"][k + 1] + " "
    elif mode == "source": 
        random_src = parallel_ds["src"][k] + " "
    elif mode == "target": 
        random_ref = parallel_ds["ref"][k + 1] + " "
    # Store translations
    translated_texts = []

    # Process dataset in batches
    batch_size = 4
    for i in range(0, len(parallel_ds), batch_size):
        batch = {key: parallel_ds[key][i:i + batch_size] for key in parallel_ds.column_names}

        # Create prompts
        prompts = batch["prompt"]
        updated_prompts = [random_src + random_ref + prompt for prompt in prompts]
        
        # Tokenization
        input_ids = tokenizer(updated_prompts, return_tensors="pt", padding=True)["input_ids"]
        ref_toks = tokenizer(batch["ref"], return_tensors="pt", padding=True)["input_ids"]

        max_len = ref_toks.shape[1]
        len_prompt = input_ids.shape[1]
        
        input_ids = input_ids.to(model.device)

        # Generate translations
        with torch.no_grad():
            output_tokens = model.generate(input_ids, max_new_tokens=max_len, num_beams=2, early_stopping=True)

        # Decode and store results
        outputs = [tokenizer.decode(output_tokens[j][len_prompt:]) for j in range(len(prompts))]
        outputs = [s.replace('</s>', '') for s in outputs]
        translated_texts.extend(outputs)
        print ("batch ", i, outputs)
    # Add translations to dataset
    parallel_ds = parallel_ds.add_column("mt", translated_texts)

    # Evaluate with COMET
    with torch.no_grad():
        scores.append(comet_model.predict(parallel_ds, batch_size=16))

    return scores
```

```python
model = AutoModelForCausalLM.from_pretrained(model_ids[3],
                                             quantization_config=quant_config,
                                             device_map='auto')
tokenizer = AutoTokenizer.from_pretrained(model_ids[3], trust_remote_code=True)
```

```python
#tokenizer.pad_token = tokenizer.eos_token
```

```python
scores_none = eval_comet_monolingual(ds.take(50), model, tokenizer, comet_model)
scores_source = eval_comet_monolingual(ds.take(50), model, tokenizer, comet_model, mode="source")
scores_target = eval_comet_monolingual(ds.take(50), model, tokenizer, comet_model, mode="target")
scores_both = eval_comet_monolingual(ds.take(50), model, tokenizer, comet_model, mode="both")
```

```python
print(scores_none, scores_source, scores_target, scores_both)
```
